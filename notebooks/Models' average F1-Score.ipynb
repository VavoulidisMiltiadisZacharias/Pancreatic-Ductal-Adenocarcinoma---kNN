### Store the average evaluation F1-Score

```python
# Store the average evaluation F1-Score
averageEvaluationF1 = {
    "k=3 - Version 1": 0.9918,
    "k=3 - Version 2": 0.9040,
    "k=3 - Version 3": 0.8583,
    "k=3 - Version 4": 0.6271,
    "k=5 - Version 1": 0.9906,
    "k=5 - Version 2": 0.9483,
    "k=5 - Version 3": 0.8488,
    "k=5 - Version 4": 0.6094,
    "k=9 - Version 1": 0.9906,
    "k=9 - Version 2": 0.9523,
    "k=9 - Version 3": 0.8149,
    "k=9 - Version 4": 0.6070,
    "k=11 - Version 1": 0.9906,
    "k=11 - Version 2": 0.9510,
    "k=11 - Version 3": 0.8013,
    "k=11 - Version 4": 0.6004,
}
```

### Extract names and values

```python
# Extract names and values
averageEvaluationF1Names = list(averageEvaluationF1.keys())
averageEvaluationF1Values = list(averageEvaluationF1.values())
```

### Plot average evaluation F1-Score

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(averageEvaluationF1Names, averageEvaluationF1Values, color='black', width=0.1)
plt.ylabel("Average evaluation F1-Score", fontsize = 10)
plt.title("Models' average evaluation F1-Score", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 1.4)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.05, f'{height:.4f}',
             ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.show()
```

### Sort average evaluation F1-Score in ascending order

```python
# Sort by values (ascending)
sorted_averageEvaluationF1 = dict(sorted(averageEvaluationF1.items(), key=lambda item: item[1]))
```

### Extract sorted names and values

```python
# Extract sorted names and values
sorted_averageEvaluationF1Names = list(sorted_averageEvaluationF1.keys())
sorted_averageEvaluationF1Values = list(sorted_averageEvaluationF1.values())
```

### Plot sorted average evaluation F1-Score

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(sorted_averageEvaluationF1Names, sorted_averageEvaluationF1Values, color='black', width=0.1)
plt.ylabel("Average evaluation F1-Score", fontsize = 10)
plt.title("Models' average evaluation F1-Score - Sorted", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 1.4)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.05, f'{height:.4f}',
             ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.show()
```
