## - General description of the "Models' execution time - Plots" file:
    
The following analysis, just plots every model's execution time with respect to version (for comparison reasons)

##############################################################################################################################

###  Import some useful libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

### Suppress warnings

```python
# import warnings

# # To suppress all warnings
# warnings.filterwarnings("ignore")
```

### Store the execution times

```python
# Store the execution times
executionTimes = {
    "k=3 - Version 1": 1.20,
    "k=3 - Version 2": 22.99,
    "k=3 - Version 3": 21.91,
    "k=3 - Version 4": 22.60,
    "k=5 - Version 1": 1.32,
    "k=5 - Version 2": 21.22,
    "k=5 - Version 3": 21.46,
    "k=5 - Version 4": 21.07,
    "k=9 - Version 1": 1.23,
    "k=9 - Version 2": 21.61,
    "k=9 - Version 3": 21.35,
    "k=9 - Version 4": 20.94,
    "k=11 - Version 1": 1.22,
    "k=11 - Version 2": 22.59,
    "k=11 - Version 3": 20.68,
    "k=11 - Version 4": 21.04,
}
```

### Extract names and values

```python
# Extract names and values
executionTimesNames = list(executionTimes.keys())
executionTimesValues = list(executionTimes.values())
```

### Plot execution times

```python
# Plot the training barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(executionTimesNames, executionTimesValues, color='black', width=0.1)
plt.ylabel("Execution time (seconds)", fontsize = 10)
plt.title("Models' execution time", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 30)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.2f}',
             ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.show()
```

### Sort the execution times in ascending order

```python
# Sort by values (ascending)
sorted_executionTimes = dict(sorted(executionTimes.items(), key=lambda item: item[1]))
```

### Extract sorted names and values

```python
# Extract sorted names and values
sorted_executionTimesNames = list(sorted_executionTimes.keys())
sorted_executionTimesValues = list(sorted_executionTimes.values())
```

### Plot sorted execution times

```python
# Plot the training barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(sorted_executionTimesNames, sorted_executionTimesValues, color='black', width=0.1)
plt.ylabel("Execution time (seconds)", fontsize = 10)
plt.title("Models' execution time (sorted)", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 30)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.2f}',
             ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.show()
```
