## <div align="center"><u> Average True Positives </u>

### Store the average True Positives 

```python
# Store the average True Positives
averageTruePositives = {
    "k=3 - Version 1": 24.60,
    "k=3 - Version 2": 22.60,
    "k=3 - Version 3": 18.88,
    "k=3 - Version 4": 11.60,
    "k=5 - Version 1": 24.54,
    "k=5 - Version 2": 22.58,
    "k=5 - Version 3": 18.50,
    "k=5 - Version 4": 11.12,
    "k=9 - Version 1": 24.54,
    "k=9 - Version 2": 22.76,
    "k=9 - Version 3": 17.26,
    "k=9 - Version 4": 11.06,
    "k=11 - Version 1": 24.54,
    "k=11 - Version 2": 22.70,
    "k=11 - Version 3": 16.80,
    "k=11 - Version 4": 10.88,
}
```

### Extract names and values

```python
# Extract names and values
averageTruePositivesNames = list(averageTruePositives.keys())
averageTruePositivesValues = list(averageTruePositives.values())
```

### Plot average True Positives 

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(averageTruePositivesNames, averageTruePositivesValues, color='black', width=0.1)
plt.ylabel("Average True Positives", fontsize = 10)
plt.title("Models' average True Positives (confusion matrix)", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 30)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.2f}',
             ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.show()
```

### Sort average True Positives in ascending order

```python
# Sort by values (ascending)
sorted_averageTruePositives = dict(sorted(averageTruePositives.items(), key=lambda item: item[1]))
```

### Extract sorted names and values

```python
# Extract sorted names and values
sorted_averageTruePositivesNames = list(sorted_averageTruePositives.keys())
sorted_averageTruePositivesValues = list(sorted_averageTruePositives.values())
```

### Plot sorted average True Positives

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(sorted_averageTruePositivesNames, sorted_averageTruePositivesValues, color='black', width=0.1)
plt.ylabel("Average True Positives", fontsize = 10)
plt.title("Models' average True Positives (confusion matrix) - Sorted", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 30)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.2f}',
             ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.show()
```
