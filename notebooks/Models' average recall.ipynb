## <div align="center"><u> Average evaluation recall </u>

### Store the average evaluation recall

```python
# Store the average evaluation recall
averageEvaluationRecall = {
    "k=3 - Version 1": 0.9840,
    "k=3 - Version 2": 0.9040,
    "k=3 - Version 3": 0.7552,
    "k=3 - Version 4": 0.4640,
    "k=5 - Version 1": 0.9816,
    "k=5 - Version 2": 0.9032,
    "k=5 - Version 3": 0.7400,
    "k=5 - Version 4": 0.4448,
    "k=9 - Version 1": 0.9816,
    "k=9 - Version 2": 0.9104,
    "k=9 - Version 3": 0.6904,
    "k=9 - Version 4": 0.4424,
    "k=11 - Version 1": 0.9816,
    "k=11 - Version 2": 0.9080,
    "k=11 - Version 3": 0.6720,
    "k=11 - Version 4": 0.4352,
}
```

### Extract names and values

```python
# Extract names and values
averageEvaluationRecallNames = list(averageEvaluationRecall.keys())
averageEvaluationRecallValues = list(averageEvaluationRecall.values())
```

### Plot average evaluation recall

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(averageEvaluationRecallNames, averageEvaluationRecallValues, color='black', width=0.1)
plt.ylabel("Average evaluation recall", fontsize = 10)
plt.title("Models' average evaluation recall", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 1.4)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{height:.4f}',
             ha='center', va='bottom', fontsize=8, rotation=90)


plt.tight_layout()
plt.show()
```

### Sort average evaluation precision in ascending order

```python
# Sort by values (ascending)
sorted_averageEvaluationRecall = dict(sorted(averageEvaluationRecall.items(), key=lambda item: item[1]))
```

### Extract sorted names and values

```python
# Extract sorted names and values
sorted_averageEvaluationRecallNames = list(sorted_averageEvaluationRecall.keys())
sorted_averageEvaluationRecallValues = list(sorted_averageEvaluationRecall.values())
```

### Plot sorted average evaluation recall

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(sorted_averageEvaluationRecallNames, sorted_averageEvaluationRecallValues, color='black', width=0.1)
plt.ylabel("Average evaluation recall", fontsize = 10)
plt.title("Models' average evaluation recall - Sorted", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 1.4)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{height:.4f}',
             ha='center', va='bottom', fontsize=8, rotation=90)


plt.tight_layout()
plt.show()
```
