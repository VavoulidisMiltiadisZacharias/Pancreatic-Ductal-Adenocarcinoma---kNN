## <div align="center"><u> Average evaluation precision </u>

### Store the average evaluation precision

```python
# Store the average evaluation precision
averageEvaluationPrecision = {
    "k=3 - Version 1": 1.0000,
    "k=3 - Version 2": 1.0000,
    "k=3 - Version 3": 1.0000,
    "k=3 - Version 4": 1.0000,
    "k=5 - Version 1": 1.0000,
    "k=5 - Version 2": 1.0000,
    "k=5 - Version 3": 1.0000,
    "k=5 - Version 4": 1.0000,
    "k=9 - Version 1": 1.0000,
    "k=9 - Version 2": 1.0000,
    "k=9 - Version 3": 1.0000,
    "k=9 - Version 4": 1.0000,
    "k=11 - Version 1": 1.0000,
    "k=11 - Version 2": 1.0000,
    "k=11 - Version 3": 1.0000,
    "k=11 - Version 4": 1.0000,
}
```

### Extract names and values

```python
# Extract names and values
averageEvaluationPrecisionNames = list(averageEvaluationPrecision.keys())
averageEvaluationPrecisionValues = list(averageEvaluationPrecision.values())
```

### Plot average evaluation precision

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(averageEvaluationPrecisionNames, averageEvaluationPrecisionValues, color='black', width=0.1)
plt.ylabel("Average evaluation precision", fontsize = 10)
plt.title("Models' average evaluation precision", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 1.4)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{height:.4f}',
             ha='center', va='bottom', fontsize=8, rotation=90)


plt.tight_layout()
plt.show()
```

### Sort average evaluation precision in ascending order

```python
# Sort by values (ascending)
sorted_averageEvaluationPrecision = dict(sorted(averageEvaluationPrecision.items(), key=lambda item: item[1]))
```

### Extract sorted names and values

```python
# Extract sorted names and values
sorted_averageEvaluationPrecisionNames = list(sorted_averageEvaluationPrecision.keys())
sorted_averageEvaluationPrecisionValues = list(sorted_averageEvaluationPrecision.values())
```

### Plot sorted average evaluation precision

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(sorted_averageEvaluationPrecisionNames, sorted_averageEvaluationPrecisionValues, color='black', width=0.1)
plt.ylabel("Average evaluation precision", fontsize = 10)
plt.title("Models' average evaluation precision - Sorted", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 1.4)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{height:.4f}',
             ha='center', va='bottom', fontsize=8, rotation=90)


plt.tight_layout()
plt.show()
```
