## <div align="center"><u> Average evaluation AUC-ROC </u>

### Store the average evaluation AUC-ROC

```python
# Store the average evaluation AUC-ROC
averageEvaluationAUCROC = {
    "k=3 - Version 1": 0.9968,
    "k=3 - Version 2": 0.9784,
    "k=3 - Version 3": 0.9584,
    "k=3 - Version 4": 0.8904,
    "k=5 - Version 1": 0.9992,
    "k=5 - Version 2": 0.9836,
    "k=5 - Version 3": 0.9776,
    "k=5 - Version 4": 0.9411,
    "k=9 - Version 1": 0.9996,
    "k=9 - Version 2": 0.9868,
    "k=9 - Version 3": 0.9912,
    "k=9 - Version 4": 0.9799,
    "k=11 - Version 1": 0.9996,
    "k=11 - Version 2": 0.9880,
    "k=11 - Version 3": 0.9972,
    "k=11 - Version 4": 0.9932,
}
```

### Extract names and values

```python
# Extract names and values
averageEvaluationAUCROCNames = list(averageEvaluationAUCROC.keys())
averageEvaluationAUCROCValues = list(averageEvaluationAUCROC.values())
```

### Plot average evaluation AUC-ROC

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(averageEvaluationAUCROCNames, averageEvaluationAUCROCValues, color='black', width=0.1)
plt.ylabel("Average evaluation AUC-ROC", fontsize = 10)
plt.title("Models' average evaluation AUC-ROC", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 1.4)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.05, f'{height:.4f}',
             ha='center', va='bottom', fontsize=8, rotation=90)


plt.tight_layout()
plt.show()
```

### Sort average evaluation AUC-ROC in ascending order

```python
# Sort by values (ascending)
sorted_averageEvaluationAUCROC = dict(sorted(averageEvaluationAUCROC.items(), key=lambda item: item[1]))
```

### Extract sorted names and values

```python
# Extract sorted names and values
sorted_averageEvaluationAUCROCNames = list(sorted_averageEvaluationAUCROC.keys())
sorted_averageEvaluationAUCROCValues = list(sorted_averageEvaluationAUCROC.values())
```

### Plot sorted average evaluation AUC-ROC

```python
# Plot the barchart
plt.figure(figsize=(8, 6))
bars = plt.bar(sorted_averageEvaluationAUCROCNames, sorted_averageEvaluationAUCROCValues, color='black', width=0.1)
plt.ylabel("Average evaluation AUC-ROC", fontsize = 10)
plt.title("Models' average evaluation AUC-ROC - Sorted", fontsize = 12)
plt.xticks(rotation=90, ha='right', fontsize = 10)  # Rotate labels for better readability

# Set y-axis limit (adjust max as needed to fit text)
plt.ylim(0, 1.4)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.05, f'{height:.4f}',
             ha='center', va='bottom', fontsize=8, rotation=90)


plt.tight_layout()
plt.show()
```
